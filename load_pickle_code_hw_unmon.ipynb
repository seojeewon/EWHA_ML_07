```python
import pickle

TOTAL_URLS = 10000  # total number in the dataset

# Load 10,000 unmon pickle file
print("Loading datafile...")
with open('/content/drive/MyDrive/Colab Notebooks/2023_2_ML/unmon_standard10.pkl', 'rb') as f:  # Path to unmon_standard10.pkl in Colab
    x = pickle.load(f)

size = len(x)
print(f'Total samples: {size}')

X1 = [] # Array to store instances (timestamps) - 10,000 instances, e.g., [[0.0, 0.5, 3.4, ...], [0.0, 4.5, ...], [0.0, 1.5, ...], ... [... ,45.8]]
X2 = [] # Array to store instances (direction*size) - size information

for i in range(TOTAL_URLS):
    size_seq = []
    time_seq = []
    for c in x[i]:
        dr = 1 if c > 0 else -1
        time_seq.append(abs(c))
        size_seq.append(dr) # In the pickle file, there is no size information, so the conversion code is set to multiply by 512 uniformly.
    X1.append(time_seq)
    X2.append(size_seq)

print(len(X1)) # Print the length of X1
```

```python
total_pkt_per_user_unmon = [] # total number of packets per user (both directions)
total_outgoing_pkt_per_user_unmon  = [] # total number of outgoing packets per user
total_incoming_pkt_per_user_unmon  = [] # total number of incoming packets per use

for i in range(10000):
  total_pkt_per_user_unmon.append(len(X2[i]))
  total_outgoing_pkt_per_user_unmon.append(X2[i].count(1))
  total_incoming_pkt_per_user_unmon.append(X2[i].count(-1))

print(total_pkt_per_user_unmon)
print(total_outgoing_pkt_per_user_unmon)
print(total_incoming_pkt_per_user_unmon)
```

```python
import tempfile
import shutil

def add_data( data, k):
    # 임시 파일 생성
    temp_file = tempfile.NamedTemporaryFile(mode='w', delete=False)

    # unmon 파일 열기 및 특정 행의 데이터 수정
    with open(csv_file_unmon, mode='r') as file:
        reader = csv.reader(file)
        writer = csv.writer(temp_file)
        for i, row in enumerate(reader):
            if i == k:  # k번째 행에 데이터 추가
                row.extend(data)
            writer.writerow(row)

    # 임시 파일을 원본 파일로 이동
    shutil.move(temp_file.name, csv_file_unmon)
```

```python
outgoing_pkt_frac_per_user_unmon = []  # number of outgoing packets as a fraction of the total number of packets
incoming_pkt_frac_per_user_unmon = []  # number of incoming packets as a fraction of the total number of packets

for i in range(10000):
    outgoing_fraction = total_outgoing_pkt_per_user_unmon[i] / total_pkt_per_user_unmon[i]
    incoming_fraction = total_incoming_pkt_per_user_unmon[i] / total_pkt_per_user_unmon[i]

    outgoing_pkt_frac_per_user_unmon.append(round(outgoing_fraction, 4))
    incoming_pkt_frac_per_user_unmon.append(round(incoming_fraction, 4))

print(outgoing_pkt_frac_per_user_unmon)
print(incoming_pkt_frac_per_user_unmon)
```

```python
outgoing30_per_user_unmon = [] # number of outgoing packets
incoming30_per_user_unmon = [] # number of incoming packets

for i in range(10000):
  cntOutgoing30 = 0
  cntIncoming30 = 0

  for j in range(30):
    if X2[i][j] == -1:
      cntOutgoing30 += 1
    else:
      cntIncoming30 += 1

  outgoing30_per_user_unmon.append(cntOutgoing30)
  incoming30_per_user_unmon.append(cntIncoming30)

print(outgoing30_per_user_unmon)
print(incoming30_per_user_unmon)
```

```python
# the sequence of bursts
import numpy as np
bursts = [[] for i in range(10000)]

avg_total_burst_per_user_unmon = []
std_total_burst_per_user_unmon = []
max_total_burst_per_user_unmon = []

avg_incoming_burst_per_user_unmon = []
std_incoming_burst_per_user_unmon = []
min_incoming_burst_per_user_unmon = []

avg_outgoing_burst_per_user_unmon = []
std_outgoing_burst_per_user_unmon = []
max_outgoing_burst_per_user_unmon = []

for i in range(10000):
  index = 0
  size = X2[i][0] # 1 or -1

  total_burst = []
  incoming_burst = []
  outgoing_burst = []

  for j in range(1, len(X2[i])):
    if X2[i][j-1] != X2[i][j]: # direction change
      total_burst.append((j-index))
      if(size > 0) :
        outgoing_burst.append(j-index)
      else :
        incoming_burst.append(index-j)
      index = j
      size = X2[i][j]

  total_burst_np = np.array(total_burst)
  incoming_burst_np = np.array(incoming_burst)
  outgoing_burst_np = np.array(outgoing_burst)

  avg_total = np.mean(total_burst)
  std_total = np.std(total_burst)
  max_total = np.max(total_burst)

  avg_incoming = np.mean(incoming_burst)
  std_incoming = np.std(incoming_burst)
  min_incoming = np.min(incoming_burst)

  avg_outgoing = np.mean(outgoing_burst)
  std_outgoing = np.std(outgoing_burst)
  max_outgoing = np.max(outgoing_burst)


  avg_total_burst_per_user_unmon.append(round(avg_total,4))
  std_total_burst_per_user_unmon.append(round(std_total,4))
  max_total_burst_per_user_unmon.append(max_total)

  avg_incoming_burst_per_user_unmon.append(round(avg_incoming,4))
  std_incoming_burst_per_user_unmon.append(round(std_incoming,4))
  min_incoming_burst_per_user_unmon.append(min_incoming)

  avg_outgoing_burst_per_user_unmon.append(round(avg_outgoing,4))
  std_outgoing_burst_per_user_unmon.append(round(std_outgoing,4))
  max_outgoing_burst_per_user_unmon.append(max_outgoing)
```

```python
interval_unmon = [[] for i in range(10000)]

for i in range(10000):
  for j in range(1, len(X1[i])):
    tmp = X1[i][j] - X1[i][j-1]
    interval_unmon[i].append(tmp)
```

```python
avg_interval_unmon = []
std_interval_unmon = []
max_interval_unmon = []

for arr in interval_unmon:
    avg_interval_unmon.append(round(np.mean(arr),4))
    std_interval_unmon.append(round(np.std(arr),4))
    max_interval_unmon.append(round(np.max(arr),4))

print(avg_interval_unmon)
print(std_interval_unmon)
print(max_interval_unmon)
```

```python
time_interval=1.0
packets_in_fixed_time_unmon = [[] for i in range(10000)]

for i in range(10000):
  start_time=X1[i][0]
  end_time=start_time+time_interval
  a=0
  for timestamp in X1[i]:
    if timestamp<=end_time:
      a+=1
    else:
      packets_in_fixed_time_unmon[i].append(a)
      a=1
      start_time=end_time
      end_time+=time_interval
  packets_in_fixed_time_unmon[i].append(a)

avg_pkts_in_fixed_time_unmon = []
std_pkts_in_fixed_time_unmon = []
min_pkts_in_fixed_time_unmon = []
max_pkts_in_fixed_time_unmon = []

for arr in packets_in_fixed_time_unmon:
    arr = np.array(arr)
    avg_pkts_in_fixed_time_unmon.append(round(np.mean(arr),4))
    std_pkts_in_fixed_time_unmon.append(round(np.std(arr),4))
    min_pkts_in_fixed_time_unmon.append(np.min(arr))
    max_pkts_in_fixed_time_unmon.append(np.max(arr))

# Print or use the computed statistics
print(avg_pkts_in_fixed_time_unmon)
print(std_pkts_in_fixed_time_unmon)
print(min_pkts_in_fixed_time_unmon)
print(max_pkts_in_fixed_time_unmon)
```

```python
def sum_interval_unmon(X1,X2):
  sum_intervals_continuous = [[] for i in range(10000)]

  for i in range(10000):
    start_pack=X2[i][0]
    start_time=X1[i][0]
    end_time=X1[i][0]

    for j in range(1,len(X1[i])):
      if start_pack == X2[i][j]: #512 or -512
        end_time=X1[i][j]
        continue
      else:
        end_time=X1[i][j]
        sum_intervals_continuous[i].append(round(end_time-start_time,3))
        start_time=end_time
    sum_intervals_continuous[i].append(round(end_time-start_time,3))

  return sum_intervals_continuous
```

```python
sum_intervals_continuous_unmon=sum_interval_unmon(X1,X2)

avg_sum_intervals_continuous_unmon = []
std_sum_intervals_continuous_unmon = []
max_sum_intervals_continuous_unmon = []

for arr in sum_intervals_continuous_unmon:
    # 각 1차원 배열의 평균, 표준편차, 최소값, 최대값 계산하여 배열에 추가
    avg_sum_intervals_continuous_unmon.append(round(np.mean(arr),4))
    std_sum_intervals_continuous_unmon.append(round(np.std(arr),4))
    max_sum_intervals_continuous_unmon.append(np.max(arr))

print(avg_sum_intervals_continuous_unmon)
print(std_sum_intervals_continuous_unmon)
print(max_sum_intervals_continuous_unmon)
```

```python
sum_intervals_continuous_unmon=sum_interval_unmon(X1,X2)

outgoing_pkts_time_interval_unmon=[[] for i in range(10000)]
incoming_pkts_time_interval_unmon=[[] for i in range(10000)]

for i in range(10000):
  for j in range(len(sum_intervals_continuous_unmon[i])):
    if j%2==0:
      outgoing_pkts_time_interval_unmon[i].append(sum_intervals_continuous_unmon[i][j])
    else:
      incoming_pkts_time_interval_unmon[i].append(sum_intervals_continuous_unmon[i][j])
```

```python
avg_outgoing_pkts_time_interval_unmon = []
std_outgoing_pkts_time_interval_unmon = []
max_outgoing_pkts_time_interval_unmon = []

avg_incoming_pkts_time_interval_unmon = []
std_incoming_pkts_time_interval_unmon = []
max_incoming_pkts_time_interval_unmon = []


for arr in outgoing_pkts_time_interval_unmon:
    avg_outgoing_pkts_time_interval_unmon.append(round(np.mean(arr),4))
    std_outgoing_pkts_time_interval_unmon.append(round(np.std(arr),4))
    max_outgoing_pkts_time_interval_unmon.append(np.max(arr))

for arr in incoming_pkts_time_interval_unmon:
    avg_incoming_pkts_time_interval_unmon.append(round(np.mean(arr),4))
    std_incoming_pkts_time_interval_unmon.append(round(np.std(arr),4))
    max_incoming_pkts_time_interval_unmon.append(np.max(arr))


print(avg_outgoing_pkts_time_interval_unmon)
print(std_outgoing_pkts_time_interval_unmon)
print(max_outgoing_pkts_time_interval_unmon)

print(avg_incoming_pkts_time_interval_unmon)
print(std_incoming_pkts_time_interval_unmon)
print(max_incoming_pkts_time_interval_unmon)
```
